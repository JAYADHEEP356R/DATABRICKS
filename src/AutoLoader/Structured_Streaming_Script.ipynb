### AUTO INGEST STREAMING PIPELINE

```python
df = spark.readStream.format("CloudFiles") \
    .option("CloudFiles.format","csv") \
    .option("CloudFiles.SchemaLocation","/Volumes/workspace/default/autoloads/destination/checkpoint/") \
    .option("CloudFiles.SchemaEvolutionMode","RESCUE") \
    .load("/Volumes/workspace/default/autoloads/raw_data")
```

```python
df.writeStream.format("delta") \
.option("checkpointLocation","/Volumes/workspace/default/autoloads/destination/checkpoint/") \
.outputMode("append") \
.trigger(once=True) \
.start("/Volumes/workspace/default/autoloads/destination/data/")
```

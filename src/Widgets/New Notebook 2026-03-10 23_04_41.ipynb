```python
# =========================================================
# 1. TEXT WIDGET
# =========================================================

# Create a text widget
dbutils.widgets.text("username", "guest", "Enter User Name")

# Read value
username = dbutils.widgets.get("username")

print("User Name is:", username)



# =========================================================
# 2. DROPDOWN WIDGET
# =========================================================

# Create dropdown widget
dbutils.widgets.dropdown(
    "department",
    "IT",
    ["IT", "HR", "Finance", "Marketing"],
    "Select Department"
)

# Read dropdown value
department = dbutils.widgets.get("department")

print("Selected Department:", department)



# =========================================================
# 3. COMBOBOX WIDGET
# =========================================================

# Combobox allows dropdown + custom input
dbutils.widgets.combobox(
    "city",
    "Bangalore",
    ["Bangalore", "Chennai", "Hyderabad"],
    "Select City"
)

city = dbutils.widgets.get("city")

print("Selected City:", city)



# =========================================================
# 4. MULTISELECT WIDGET
# =========================================================

# Allows selecting multiple values
dbutils.widgets.multiselect(
    "countries",
    "India",
    ["India", "USA", "UK", "Germany"],
    "Select Countries"
)

countries = dbutils.widgets.get("countries")

print("Selected Countries:", countries)

# Convert to list
country_list = countries.split(",")

print("Country List:", country_list)



# =========================================================
# 5. GET ALL WIDGET VALUES
# =========================================================

all_widgets = dbutils.widgets.getAll()

print("All Widget Values:", all_widgets)



# =========================================================
# 6. EXAMPLE USING WIDGET VALUES
# =========================================================

# Sample employee data
data = [
    (1, "John", "IT"),
    (2, "Sara", "HR"),
    (3, "Mike", "Finance"),
    (4, "David", "Marketing")
]

columns = ["id", "name", "department"]

df = spark.createDataFrame(data, columns)

df.show()

# Filter using widget value
dept = dbutils.widgets.get("department")

filtered_df = df.filter(f"department = '{dept}'")

print("Filtered Data Based on Widget:")
filtered_df.show()



# =========================================================
# 7. PASSING PARAMETERS BETWEEN NOTEBOOKS
# =========================================================

# Example only (works when another notebook exists)

# result = dbutils.notebook.run("NotebookB", 60, {"country":"India"})
# print(result)



# =========================================================
# 8. REMOVE WIDGETS
# =========================================================

# Remove a specific widget
# dbutils.widgets.remove("username")

# Remove all widgets
# dbutils.widgets.removeAll()
```

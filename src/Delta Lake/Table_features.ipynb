```python
%sql
CREATE TABLE if not exists sales (
  id INT,
  product STRING,
  price INT,
  country STRING,
  order_date DATE
)
USING DELTA;
```

```python
%sql
ALTER TABLE sales
SET TBLPROPERTIES (delta.enableChangeDataFeed = true);
```

```python
%sql
INSERT INTO sales VALUES
(1,'Shoes',1000,'India','2024-01-01'),
(2,'Watch',2000,'USA','2024-01-02'),
(3,'Bag',1500,'India','2024-01-03');
```

```python
%sql
SELECT * FROM sales VERSION AS OF 1;
```

```python
%sql
CREATE TABLE if not EXISTS sales_partitioned
USING DELTA
PARTITIONED BY (country)
AS SELECT * FROM sales;
```

```python
%sql
OPTIMIZE sales_partitioned
ZORDER BY (order_date);
```

```python
%sql
SELECT * FROM table_changes('sales', 1);
```

```python
%sql
CREATE TABLE if not EXISTS order_summary (
 price INT,
 tax INT,
 total INT GENERATED ALWAYS AS (price + tax)
)
USING DELTA;

CREATE TABLE if not EXISTS users (
 id BIGINT GENERATED ALWAYS AS IDENTITY,
 name STRING
)
USING DELTA;
```

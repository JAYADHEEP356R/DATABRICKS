```python
df = spark.read \
.format("csv") \
.option("header",True) \
.option("inferSchema",True) \
.load("/Volumes/workspace/data_files/csv_files/governence.csv")


df.write \
.format("delta") \
.mode("overwrite") \
.option("header",True) \
.saveAsTable("governance")
```

```python
%sql
SELECT * FROM governance
```

Governance Technique 3 — Row Level Security

```python
%sql
CREATE VIEW if not EXISTS customers_chennai AS
SELECT *
FROM governance
WHERE city='Chennai';
```

```python
%sql
GRANT SELECT
ON VIEW customers_chennai
TO `jayadheep356@gmail.com`;
```

```python
%sql
select * from customers_chennai
```

Governance Technique 4 — Data Masking

```python
%sql
CREATE VIEW if not EXISTS customers_masked AS
SELECT
name,
concat('****@', split(email,'@')[1]) as email,
city,
salary
FROM governance
```

```python
%sql
select * from customers_masked
```
